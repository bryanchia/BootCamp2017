```python
#Problem 2

def plot2(x):
    
    cosx = np.cos(x)
    plt.plot(x,cosx, label = 'cosx')
    sinx = np.sin(x)
    plt.plot(x,sinx, label = 'sinx')
    arctanx = np.arctan(x)
    plt.plot(x,arctanx, label = 'arctanx')
    plt.legend(loc='upper left')
    plt.ylim([-3, 3])
    plt.show()
    
    return

x = np.linspace(-2*np.pi, 2*np.pi, 100)
plot2(x)
```

```python
#Problem 3

def plot3(x):
    
    plt.clf()
    
    x1 = x[x < 1]
    x2 = x[x > 1]
    
    y1 = 1 / (x1 - 1)
    y2 = 1 / (x2 - 1)
    
    
    plt.plot(x1, y1, 'm--', linewidth=4)
    plt.plot(x2, y2, 'm--', linewidth=4)
    plt.ylim([-6, 6])
    plt.show()
    return

x = np.linspace(-2, 6, 100)
plot3(x)
```

```python
#Problem 4

def plot4(x):
    
    plt.clf()
    
    fig, axes = plt.subplots(nrows=2, ncols=2)
    
    sinx = np.sin(x)
    plt.subplot(221)
    plt.plot(x,sinx,'g-')
    plt.title('sinx')
    plt.axis([0, 2*np.pi,-2,2])
    
    sin2x = np.sin(2*x)
    plt.subplot(222)
    plt.plot(x,sin2x,'r-')
    plt.title('sin2x')
    plt.axis([0, 2*np.pi,-2,2])

    twosinx = 2 * np.sin(x)
    plt.subplot(223)
    plt.plot(x,twosinx,'b--')
    plt.title('2sinx')
    plt.axis([0, 2*np.pi,-2,2])

    twosin2x = 2 * np.sin(2*x)
    plt.subplot(224)
    plt.plot(x,twosin2x,'m.')
    plt.title('2sin2x')
    plt.axis([0, 2*np.pi,-2,2])
    
    fig.subplots_adjust(hspace=0.5)
    fig.suptitle("Sine Function Graphs")
    plt.show()
    
    return

x = np.linspace(0, 2*np.pi, 100)
plot4(x)
```

```python
#Problem 5

def visualize():
    
    plt.clf()
    fig, axes = plt.subplots(nrows=2, ncols=1)
    fig.set_figheight(10)
    fig.set_figwidth(10)
    
    data = np.load("FARS.npy")
    
    long = np.array(data[:,1])
    lat = np.array(data[:,2])
    plt.subplot(211)
    plt.plot(long,lat, 'ko', markersize=0.03)
    plt.axis("equal")
    
    hour = np.array(data[:,0])
    plt.subplot(212)
    plt.hist(hour, bins=np.arange(-1, 25))
    plt.xlabel("Hour of the day")
    
    fig.subplots_adjust(hspace=1)
    plt.show()
    return
    
visualize()
```

```python
#Problem 6

def plot6():
    
    plt.clf()
    x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
    y = x.copy()
    X, Y = np.meshgrid(x, y)
    
    fig, axes = plt.subplots(nrows=2, ncols=1)
    fig.set_figheight(10)
    fig.set_figwidth(10)
    
    Z = (np.sin(X) * np.sin(Y)) / (X * Y)
    
    plt.subplot(211)
    plt.pcolormesh(X, Y, Z, cmap="viridis")
    plt.colorbar()
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-2 * np.pi, 2 * np.pi)
    
    plt.subplot(212)
    plt.contour(X, Y, Z, 15, cmap="Spectral")
    plt.colorbar()
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-2 * np.pi, 2 * np.pi)
    
    fig.subplots_adjust(hspace=1)
    plt.show()
    
    return

plot6()
```
