```python
#Problem 1

import pandas as pd
import numpy as np
import pickle

acc = pd.read_pickle("final_accidents2.pickle")

#acc = df2[(df2['LATITUDE'] != 77.7777) & (df2['LATITUDE'] != 88.8888) & (df2['LATITUDE'] != 99.9999) & (df2['LONGITUD'] != 777.7777) & (df2['LONGITUD'] != 888.8888) & (df2['LONGITUD'] != 999.9999)]

acc = acc.rename(columns={'SP': 'SPEEDING'})

from pyproj import Proj, transform

from_proj = Proj(init="epsg:4326")
to_proj = Proj(init="epsg:3857")

def convert(longitudes, latitudes):
    """Converts latlon coordinates to meters.
    Inputs:
    longitudes (array-like) : array of longitudes
    latitudes (array-like) : array of latitudes
    Example:
    x,y = convert(accidents.LONGITUD, accidents.LATITUDE)
    """
    x_vals = []
    y_vals = []
    for lon, lat in zip(longitudes, latitudes):
        x, y = transform(from_proj, to_proj, lon, lat)
        x_vals.append(x)
        y_vals.append(y)
    return x_vals, y_vals

acc['LONGITUD'], acc['LATITUDE'] = convert(acc['LONGITUD'], acc['LATITUDE'])

acc
```

```python
#Problem 6

#Split data up into three types 

df_drunk = acc[acc["DRUNK_DR"]==1]

df_speed = acc[acc["SPEEDING"]==1]

df_other = acc[(acc["SPEEDING"]==0) & (acc["DRUNK_DR"]==0)]

#Plot drunk drivers

df1 = pd.DataFrame({"x_vals":df_drunk["LONGITUD"],
"y_vals":df_drunk["LATITUDE"]})

df1["fill_alpha"] = 1
df1["fill_color"] = "red"
df1["line_alpha"] = 0
df1["size"] = 1

cir_source = ColumnDataSource(df1)

cir = fig.circle(x="x_vals", y="y_vals", source=cir_source, radius="size", fill_color="fill_color", fill_alpha="fill_alpha", line_width=0)

#Plot speeding drivers

df2 = pd.DataFrame({"x_vals":df_speed["LONGITUD"],
"y_vals":df_speed["LATITUDE"]})

df2["fill_alpha"] = 1
df2["fill_color"] = "green"
df2["line_alpha"] = 0
df2["size"] = 1

cir_source2 = ColumnDataSource(df2)

cir2 = fig.circle(x="x_vals", y="y_vals", source=cir_source2, radius="size", fill_color="fill_color", fill_alpha="fill_alpha",line_width=0)

#Plot other drivers 

df3 = pd.DataFrame({"x_vals":df_other["LONGITUD"],
"y_vals":df_other["LATITUDE"]})

df3["fill_alpha"] = 1
df3["fill_color"] = "blue"
df3["line_alpha"] = 0
df3["size"] = 1

cir_source3 = ColumnDataSource(df3)

cir3 = fig.circle(x="x_vals", y="y_vals", source=cir_source3, radius="size", fill_color="fill_color", fill_alpha="fill_alpha", line_width=0)

show(fig)
```
