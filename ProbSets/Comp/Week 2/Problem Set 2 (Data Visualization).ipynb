```python
#Problem 1

import numpy as np
from matplotlib import pyplot as plt
import pylab

pylab.rcParams['figure.figsize'] = (10, 10)

data = np.load("anscombe.npy")

x1 = np.array(data[:,0])
y1 = np.array(data[:,1])

x2 = np.array(data[:,2])
y2 = np.array(data[:,3])

x3 = np.array(data[:,4])
y3 = np.array(data[:,5])

x4 = np.array(data[:,6])
y4 = np.array(data[:,7])

plt.clf()
fig, axes = plt.subplots(nrows=4, ncols=4)
xreg = np.arange(0,20)
yreg = 0.5 * xreg + 3


plt.subplot(221)
plt.plot(x1,y1, 'o', markersize=5)
plt.plot(xreg,yreg)
plt.title("Section I")
plt.xlim([0, 20])
plt.ylim([0, 20])

plt.subplot(222)
plt.plot(x2,y2, 'o', markersize=5)
plt.plot(xreg,yreg)
plt.title("Section II")
plt.xlim([0, 20])
plt.ylim([0, 20])

plt.subplot(223)
plt.plot(x3,y3, 'o', markersize=5)
plt.plot(xreg,yreg)
plt.title("Section III")
plt.xlim([0, 20])
plt.ylim([0, 20])

plt.subplot(224)
plt.plot(x4,y4, 'o', markersize=5)
plt.plot(xreg,yreg)
plt.title("Section IV")
plt.xlim([0, 20])
plt.ylim([0, 20])

fig.subplots_adjust(hspace=0.5)
plt.show()

print("The first section seems best fitted by a linear regression line.")
print("The second section seems best fitted by a quadratic equation.")
print("The third section seems best fitted by a linear regression line but the values are a\nlot tighter with the regression line compared to the first section except for one outlier in a y value.")
print("The fourth section has very little variance in x values except for one huge outlier.")
```

```python
#Problem 2

from scipy.special import comb

pylab.rcParams['figure.figsize'] = (15, 15)

x = np.linspace(0,1,50)

plt.clf()

for n in range(0,4):
        
    for v in range (0, n+1):
            
        plt.subplot(4, 4 , v + 4*n + 1)
        b = comb(n,v) * x**v * (1 - x)**(n-v)
        plt.plot(x, b, lw=2)
        title = "n="+str(n)+", v="+str(v)
        plt.title(title)
        plt.axis([0, 1.1, 0, 1.1])
    
        plt.tick_params(which="both", top="off", right="off")
        if v + 4*n + 1  < 12: # Remove x-axis label on upper plots.
            plt.tick_params(labelbottom="off")
        if (v + 4*n + 1) % 4 != 1: # Remove y-axis label on right plots.
            plt.tick_params(labelleft="off")
        
fig.subplots_adjust(hspace=0.5)
plt.show()
```

```python
#Problem 3

from scipy import stats

pylab.rcParams['figure.figsize'] = (10, 20)

data = np.load("MLB.npy")
height = np.array(data[:,0])
weight = np.array(data[:,1])
ages = np.array(data[:,2])

plt.subplot(311) 
plt.scatter(weight, height, c=ages, s=20)
fit = np.polyfit(weight, height, deg=1)
plt.plot(weight, fit[0] * weight + fit[1], color='red')
cbar = plt.colorbar()
cbar.set_label("Ages(years)")
plt.grid()
plt.xlabel("Weight(pounds)")
plt.ylabel("Height(inches)")

plt.subplot(312) 
plt.scatter(weight, ages, c=height, s=20)
fit = np.polyfit(weight, ages, deg=1)
plt.plot(weight, fit[0] * weight + fit[1], color='red')
cbar = plt.colorbar()
cbar.set_label("Height(inches)")
plt.grid()
plt.xlabel("Weight(pounds)")
plt.ylabel("Ages(years)")

plt.subplot(313) 
plt.scatter(ages, height, c=weight, s=20)
fit = np.polyfit(ages, height, deg=1)
plt.plot(ages, fit[0] * ages + fit[1], color='red')
cbar = plt.colorbar()
cbar.set_label("Weight(pounds)")
plt.grid()
plt.xlabel("Ages(years)")
plt.ylabel("Height(inches)")

fig.subplots_adjust(hspace=0.5)
plt.show()

print("There is a strong correlation between height and weight.")
print("There is a milder correlation between age and weight.")
print("There is almost no correlation between age and height.")
```

```python
#Problem 5

from matplotlib.colors import LogNorm

pylab.rcParams['figure.figsize'] = (10, 15)

x = np.linspace(-20, 20, 200)
X, Y = np.meshgrid(x, x.copy())
Z = (1-X)**2 + 100 * (Y-X**2)**2

plt.subplot(211) # Plot a heat map of f.
plt.pcolormesh(X, Y, Z, cmap="viridis")
plt.plot([1], [1], marker='o', markersize=7, color="red")
plt.colorbar()

plt.subplot(212) # Plot a contour map with 6 level curves.
plt.contour(X, Y, Z, 12, cmap="viridis", norm=LogNorm())
plt.plot([1], [1], marker='o', markersize=7, color="red")
plt.colorbar()

plt.show()
```
