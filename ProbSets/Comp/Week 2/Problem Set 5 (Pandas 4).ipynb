```python
import pandas as pd
from datetime import datetime
import numpy as np

#Problem 1

df = pd.read_csv('DJIA.csv', delimiter =",", header = 0, index_col = 0)

df.index = pd.to_datetime(df.index)

df[df['VALUE']=="."] = np.nan

df['VALUE'] = df['VALUE'].astype(float)
```

```python
#Problem 2

fri1 = pd.date_range(start='3/13/2008', periods = int(93/2) + 1, freq="WOM-1FRI")
fri2 = pd.date_range(start='3/13/2008', periods = int(93/2) , freq="WOM-3FRI" )

dates = fri1.union(fri2)

df2 = pd.read_csv('paychecks.csv', delimiter =",",names = ['paycheck'],  header = None, index_col = False)

df3 = df2.set_index(dates)

df3
```

```python
#Problem 5: 

df5 = pd.read_csv('website_traffic.csv', delimiter =",",  header = 0, index_col = False)

df5["ENTER"] = pd.to_datetime(df5["ENTER"])
df5["LEAVE"] = pd.to_datetime(df5["LEAVE"])
df5["DURATION"] = df5["LEAVE"] - df5["ENTER"]

p = pd.to_datetime(df5["ENTER"])

df51 = df5.set_index(p)

df51['DURATION'] = df51['DURATION'].apply(lambda x: pd.to_timedelta(x))
df51['DURATION'] = df51['DURATION'].apply(lambda x: x / pd.offsets.Minute(1))

df52 = df51.resample('D').mean()

print("Daily Mean Visit Duration (In Minutes)")
print(df52)
print("\n")

df51['COUNT'] = 1
df53 = df51.resample('H').sum()

print("Daily Number of Visits (By Hour)")
print(df53)
```

```python
#Problem 6

df_1 = df.shift(1)
df_change = df_1 - df

print("Day with largest gain:")
print(df_change.loc[df_change['VALUE'].idxmax()])
print("\n")
print("Day with largest loss")
print(df_change.loc[df_change['VALUE'].idxmin()])
print("\n")

df_month = df.resample('M').mean()
df_month.index = df_month.index.to_period("M")
df_month_change = df_month.shift(1) - df_month

print("Month with largest gain")
print("\n")
print(df_month_change.loc[df_change['VALUE'].idxmax()])
print("Month with largest loss")
print("\n")
print(df_month_change.loc[df_change['VALUE'].idxmax()])
```

```python
#Problem 7

import matplotlib.pyplot as plt
import pylab

pylab.rcParams['figure.figsize'] = (15, 10)
plt.style.use('ggplot')

df7 = df
df7['ROLL_30'] = df7['VALUE'].rolling(window=30).mean() #pd.rolling_mean(df7['VALUE'],30)
df7['ROLL_365'] = df7['VALUE'].rolling(window=365).mean()
df7['EXP_30'] = df7['VALUE'].ewm(span=30).mean()
df7['EXP_365'] = df7['VALUE'].ewm(span=365).mean()

df7.plot()
plt.legend(["Original", "Rolling(30)", "Rolling(365)", "EWMP(30)", "EWMP(365)"], fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Date", fontsize = 20)
plt.ylabel("Value", fontsize = 20)
plt.title("Rolling Average and EWMA", fontsize = 25)
plt.show()
```
