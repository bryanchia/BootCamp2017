```python
#Examples

import pandas as pd
import numpy as np
from pydataset import data

msleep = data("msleep")

vorder = msleep.groupby(["vore", "order"])

vore = msleep.groupby("vore")

means = vore.mean()

print(means)
errors = vore.std()
print(errors)
```

```python
#Problem 1

import matplotlib.pyplot as plt

diamonds = data("diamonds")

plt.clf()

for i, group in diamonds.groupby('clarity'):
    plt.figure()
    group.plot.scatter(x='carat', y='price', title=str(i)+", Source: R Documentation", s=2)
    plt.show()

print("I wanted to see if carats and prices don't have any correlation when the clarity of")
print("the diamonds are not as clear.")
print("These graphs show that no matter what the clarity is, there is a general upward trend")
print("between the carats and the prices.")
```

```python
#Problem 2

df = pd.read_csv('titanic.csv', delimiter =",", header = 0, index_col = None)

age_mean = df["Age"].mean()

df['Age'][pd.isnull(df['Age'])] = age_mean

titanic = df[(pd.notnull(df['Age'])) & (pd.notnull(df['Sex'])) & (pd.notnull(df['Pclass'])) & (pd.notnull(df['Embarked'])) & (pd.notnull(df['Fare']))][['Survived', 'Pclass', 'Sex', 'Age', 'Fare','Embarked']]

df4 = titanic.groupby(['Embarked'])[["Survived"]].mean()

print("Groupby Statistics:")
print(df4)

df3 = titanic.pivot_table('Survived', index =['Embarked'], columns='Sex')

print("\n")
print("Pivot Table Statistics:")
print(df3)
```
